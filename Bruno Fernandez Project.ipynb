```python
fig, ax = plt.subplots(figsize=(20, 10))
pitch = VerticalPitch(
    pitch_type='statsbomb', 
    line_color='white', 
    linestyle='--', 
    linewidth=1, 
    goal_linestyle='-', 
    pitch_color="grass", 
    half=False
)
pitch.draw(ax=ax)

# Plot the shots
sc = pitch.scatter(
    bruno_shots['X'], bruno_shots['Y'], 
    s=(bruno_shots['xG'] * 1000),  # Scale xG for marker size
    c=bruno_shots['xG'],           # Color by xG
    cmap='coolwarm', 
    edgecolors='black', 
    ax=ax
)

# Add a color bar
cbar = plt.colorbar(sc, ax=ax, orientation='vertical', fraction=0.02, pad=0.02)
cbar.set_label('xG (Expected Goals)', color='black', fontsize=12)

# Add a title
ax.set_title("Bruno Fernandez Season 2024(till December 22)", color="black", fontsize=14)

plt.show()
```
